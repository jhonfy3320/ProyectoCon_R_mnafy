```r
# Instalar (si no lo tienes) y cargar librerías esenciales
install.packages(c("tidyverse", "GGally"))
library(tidyverse)   # Incluye ggplot2, dplyr, tidyr…
library(GGally)      # Funciones para pair plots

mtcars <- read.csv('/home/jhonfypy/ProyectoCon_R_mnafy/mtcars.csv')
orangeec <- read_csv("/home/jhonfypy/ProyectoCon_R_mnafy/orangeec.csv")
```

# EDA Scater plot con nuetro dataset mtcars

```r
#Exploracion de variables 
plot(mtcars$mpg ~ mtcars$cyl,
     xlab= 'Cilindros', ylab = 'Millas por galon',
     main= 'Relacion de cilindros y millaspor galon')
```

```r
plot(mtcars$mpg ~ mtcars$hp,
     xlab='Caballos de furza ', ylab = 'Millas por galon',
     main = 'Relacion caballos de fuerza y millas por galon')
```

```r
#Explorando las variables de nueto dataset orangeec
plot(orangeec$Unemployment ~ orangeec$`Education invest % GDP`,
     xlab = 'Imversion educacion (%PIB)',
     ylab = 'Desempleo',
     main = 'Relacion en educacion y desempleo ')
```

```r
plot(orangeec$`GDP PC` ~ orangeec$`Creat Ind % GDP`,
     xlab = 'Aporte economia naranja al (%PIB)',
     ylab = 'PIB per capital',
     main = 'Relacion Economia naranja pib per capital ')
```

```r
plot(mtcars$hp, mtcars$mpg,
     main = "Relación entre Caballos de Fuerza y Millas por Galón",
     xlab = "Caballos de Fuerza (hp)", 
     ylab = "Millas por Galón (mpg)", 
     col = "darkgreen", pch = 19)
```

```r
ggplot(mtcars, aes(x = hp, y = mpg)) +
  geom_point(color = "darkgreen") +
  labs(title = "Relación entre Caballos de Fuerza y Millas por Galón",
       x = "Caballos de Fuerza (hp)", 
       y = "Millas por Galón (mpg)") +
  theme_minimal()
```

```r
mtcars$am <- factor(mtcars$am, labels = c("Automática", "Manual"))
```

```r
ggplot(mtcars, aes(hp,qsec))+
    geom_point(aes(color=am, size=cyl))+
    labs(s='Caballos de fuerza', y='Tiempo en 1/4 milla',
        title ='Caballos-Velocidad segun cilindros y tipo caja')
```

```r
# Gráfico de dispersión de la penetración de internet y el crecimiento del PIB
plot(orangeec$`Internet penetration % population`, orangeec$`GDP Growth %`,
     main = "Relación entre Penetración de Internet y Crecimiento del PIB",
     xlab = "Penetración de Internet (% población)", 
     ylab = "Crecimiento del PIB (%)", 
     col = "blue", pch = 19)
```

```r
# Gráfico de dispersión con ggplot2 para orangeec
ggplot(orangeec, aes(x = `Internet penetration % population`, y = `GDP Growth %`)) +
  geom_point(color = "blue") +
  labs(title = "Relación entre Penetración de Internet y Crecimiento del PIB",
       x = "Penetración de Internet (% población)", 
       y = "Crecimiento del PIB (%)") +
  theme_minimal()
```

```r
economy <- mean(orangeec$`GDP PC`)
economy
```

```r
orangeec <- orangeec %>%
  mutate(Strong_economy = ifelse(`GDP PC` < economy,
         'Por debajo promedio pib per cápital',
         'Sobre-arriba promedio pib per cápital'))
```

```r
ggplot(orangeec, aes(`Internet penetration % population`, `Creat Ind % GDP`))+
  geom_point(aes(color = factor(Strong_economy), size = `GDP Growth %`))+
  labs(x= 'Penetración Internet', y = 'Aporte economia naranja PIB ',
       title= 'Internet y aporte conomia naranja segun economia y aporte pib')
```
