```r
# Instalar (si no lo tienes) y cargar librerías esenciales
install.packages(c("tidyverse", "GGally"))
library(tidyverse)   # Incluye ggplot2, dplyr, tidyr…
library(GGally)      # Funciones para pair plots
```

```r
mtcars <- read.csv('/home/jhonfypy/ProyectoCon_R_mnafy/mtcars.csv')
```

```r
qplot(mtcars$hp,
      geom = 'histogram',
      bins = 10,  # Puedes ajustar el número de barras
      fill = I("steelblue"), 
      color = I("black"),
      xlab = 'Caballos de fuerza',
      main = 'Distribución de Caballos de Fuerza en Carros')

```

```r
ggplot(mtcars, aes(x=hp))+
  geom_histogram()+
  labs(x= 'Caballos de furza',
       y = 'Cantidad de carros',
       title = 'Caballos de fuerza en carros seleccionados')+
  theme(panel.background = element_blank(), panel.grid.major = element_blank(),
        panel.grid.minor = element_blank())
```

```r
ggplot(mtcars, aes(x=hp))+
  geom_histogram(binwidth = 30)+
  labs(x= 'Caballos de furza',
       y = 'Cantidad de carros',
       title = 'Caballos de fuerza en carros seleccionados')+
  theme(panel.background = element_blank(), panel.grid.major = element_blank(),
        panel.grid.minor = element_blank())
```

```r
ggplot()+ geom_histogram(data=mtcars,
                         aes(x=hp), fill='blue', color='red',
                         binwidth = 20)+
  labs(x= 'Caballos de furza',
       y = 'Cantidad de carros',
       title = 'Caballos de fuerza en carros seleccionados')+
  xlim(c(80, 280))+
  theme(panel.background = element_blank(), panel.grid.major = element_blank(),
        panel.grid.minor = element_blank())
```

```r
# Histograma de Millas por Galón en mtcars
hist(mtcars$mpg,
     main = "Distribución de Millas por Galón",
     xlab = "Millas por Galón (mpg)",
     col = "lightblue",
     border = "black",
     breaks = 10)  # Ajustar la cantidad de intervalos
```

```r

# Histograma de Millas por Galón (mpg) en mtcars usando ggplot2
ggplot(mtcars, aes(x = mpg)) +
  geom_histogram(fill = "lightblue", color = "black", bins = 10) +
  labs(title = "Distribución de Millas por Galón",
       x = "Millas por Galón (mpg)", 
       y = "Frecuencia") +
  theme_minimal()
```

```r
# Histograma de millas por galón
ggplot(mtcars, aes(x = mpg)) +
  geom_histogram(binwidth = 2, fill = "skyblue", color = "black") +
  labs(title = "Distribución de MPG", x = "Millas por galón", y = "Frecuencia")
```
