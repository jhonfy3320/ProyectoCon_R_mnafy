## EDA con dataset el dataset Orangeec - box plot- ggplot2 - dplyr

```r
# Instalar (si no lo tienes) y cargar librerías esenciales
install.packages(c("tidyverse", "GGally"))
library(tidyverse)   # Incluye ggplot2, dplyr, tidyr…
library(GGally)      # Funciones para pair plots

mtcars <- read.csv('/home/jhonfypy/ProyectoCon_R_mnafy/mtcars.csv')
orangeec <- read_csv("/home/jhonfypy/ProyectoCon_R_mnafy/orangeec.csv")
```

```r
names(orangeec)
```

```r
economy <- mean(orangeec$`GDP PC`)
economy
```

```r
orangeec <- orangeec %>%
  mutate(Strong_economy = ifelse(`GDP PC` < economy,
         'Por debajo promedio pib per cápital',
         'Sobre-arriba promedio pib per cápital'))
```

```r
orangeec
```

```r
str(orangeec)
```

```r
ggplot(orangeec, aes(x= `Strong_economy`, y = `Creat Ind % GDP`,
                     fill=`Strong_economy`))+
  geom_boxplot(alpha=0.4)+                 
  labs(x= 'Tipo de pais', y = 'Aporte economia al pib',
       title = 'Aporte E.N en PIB Paises Latam con alto y bajo pib per cápital')+
  theme(legend.position = 'none') +
  theme(panel.background = element_blank(),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank())
```

```r
ggplot(orangeec, aes(x= `Strong_economy`, y = `Internet penetration % population`,
                     fill=`Strong_economy`))+
  geom_boxplot(alpha=0.4)+                 
  labs(x= 'Tipo de pais', y = 'Penetracion Internet(%)',
       title = 'Penetracion intenet en Paises Latam con alto y bajo pib per cápital')+
  theme(legend.position = 'none') +
  theme(panel.background = element_blank(),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank())
```

```r
# Calcular el promedio de la columna 'GDP PC'
economy_avg <- mean(orangeec$`GDP PC`, na.rm = TRUE)
economy_avg 
```

```r
# Crear el box plot para la variable 'Inflation'
ggplot(orangeec, aes(x = Strong_economy, y = Inflation, fill = Strong_economy)) +
  geom_boxplot(alpha = 0.4) +
  labs(x = 'Tipo de país', 
       y = 'Inflación (%)',
       title = 'Inflación en Países LATAM con Alto y Bajo PIB per Cápita') +
  theme(legend.position = 'none') +
  theme(panel.background = element_blank(),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank())
```

```r
colnames(orangeec)
str(orangeec)
colnames(orangeec)
```

```r
orangeec <- orangeec %>%
  mutate(
    Pais = as.factor(`Country`),               # ✅ con backticks (alt+96)
    Industria = as.factor(`Creat Ind % GDP`)   # ✅ también con backticks
  )
```

```r
ggplot(orangeec, aes(x = Industria, y = `GDP PC`, fill = Industria)) +
  geom_boxplot() +
  labs(
    title = "PIB per cápita por Industria Creativa",
    x = "Industria Creativa",
    y = "PIB per cápita (USD)"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1),
        legend.position = "none")

```

### ¿Qué puedes observar en estos gráficos?
