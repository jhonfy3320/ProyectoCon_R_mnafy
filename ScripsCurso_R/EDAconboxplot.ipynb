✅ ¿Qué es un boxplot en EDA?
Un boxplot muestra:

Mediana (línea dentro de la caja).

Cuartiles (Q1 y Q3, los bordes de la caja).

Rango intercuartílico (IQR = Q3 - Q1).

Bigotes (whiskers): lo que está dentro de 1.5 × IQR.

Puntos atípicos (outliers): los valores fuera de ese rango.

 ¿Qué puedes detectar en un EDA con boxplots? 1. 
**Comparación entre grupos** (ej. `am`, `cyl`, `gear`). 2. 
**Valores extremos** que podrían influir en tus modelos. 3. 
**Distribución asimétrica** en los datos. 4. 
**Diferencias de mediana**

```r
# Instalar (si no lo tienes) y cargar librerías esenciales
install.packages(c("tidyverse", "GGally"))
library(tidyverse)   # Incluye ggplot2, dplyr, tidyr…
library(GGally)      # Funciones para pair plots

mtcars <- read.csv('/home/jhonfypy/ProyectoCon_R_mnafy/mtcars.csv')
orangeec <- read_csv("/home/jhonfypy/ProyectoCon_R_mnafy/orangeec.csv")
```

```r
#Boxpot
boxplot(mtcars$hp,
        ylab= 'caballos de fuerza',
        main= 'Caballos de fuerzas en carros ')
```

```r
boxplot(mtcars$hp, mtcars$cyl,
        ylab= 'Cilindraje',
        xlab= 'caballos de fuerza',
        main = 'Cilindraje en carros ')
```

```r
boxplot(mtcars$hp, mtcars$mpg,
     main = "Relación entre Caballos de Fuerza y Millas por Galón",
     xlab = "Caballos de Fuerza (hp)", 
     ylab = "Millas por Galón (mpg)", 
     col = "darkgreen", pch = 19)
```

```r
ggplot(mtcars,aes(x=as.factor(cyl),y=hp, fill = cyl))+
  geom_boxplot(alpha = 0.8)+
  labs(x = 'Cilindros', y = 'Cabalos de fuerza',
       title = 'Cabalos de fuerza segun cilindros en mtcars')+
  theme(legend.position = 'none') +
  theme(panel.background = element_blank(),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank())
```

```r
mtcars$am <- factor(mtcars$am, labels = c("Automática", "Manual"))
```

```r
ggplot(mtcars,aes(x=am, y=mpg, fill = am))+
  geom_boxplot()+
  labs(x = 'Tipo caja', y = 'Millas por galon',
       title = 'Millas galon segun tipo caja')+
  theme(legend.position = 'none') +
  theme(panel.background = element_blank(),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank())
```

```r
library(ggplot2)

# Convertir 'am' en factor con etiquetas amigables
mtcars$am <- factor(mtcars$am, labels = c("Automática", "Manual"))

# Crear boxplot
ggplot(mtcars, aes(x = am, y = mpg, fill = am)) +
  geom_boxplot() +
  labs(
    x = "Tipo de transmisión",
    y = "Millas por galón (mpg)",
    title = "Distribución de MPG según tipo de transmisión"
  ) +
  theme_minimal() +
  theme(legend.position = "none")
```

```r
mtcars$cyl <- factor(mtcars$cyl)

ggplot(mtcars, aes(x = cyl, y = mpg, fill = am)) +
  geom_boxplot(position = position_dodge(0.8)) +
  labs(x = "Número de cilindros", y = "Millas por galón",
       title = "MPG según cilindros y tipo de transmisión",
       fill = "Transmisión") +
  theme_minimal()
```
