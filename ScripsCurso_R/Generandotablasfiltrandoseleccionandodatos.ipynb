```r
# Instalar (si no lo tienes) y cargar librerías esenciales
install.packages(c("tidyverse", "GGally"))
library(tidyverse)   # Incluye ggplot2, dplyr, tidyr…
library(GGally)      # Funciones para pair plots

mtcars <- read.csv('/home/jhonfypy/ProyectoCon_R_mnafy/mtcars.csv')
orangeec <- read_csv("/home/jhonfypy/ProyectoCon_R_mnafy/orangeec.csv")
```

```r
#Generando tablas, filtrando y seleccionando datos - dplyr-Parte 1 Dataset 
Eficientes <- mean(mtcars$mpg)
Eficientes
```

```r
#Ajustando los datos para una mejor visualizacion de los datos 
mtcars <- mtcars %>%
  mutate(Mas_eficientes=ifelse(`mpg`<Eficientes,
                               'Bajo promedio', 'en ó sobre promedio'))
```

```r
#Carros mas veloces 
Mas_veloces <- mtcars[mtcars$qsec<16,]
Mas_veloces
```

```r
mtcars <- mtcars %>%
  mutate(Velocidad_cuarto_milla=ifelse(qsec <16,
                                       'Menos 16 segundos',
                                       'Mas 16 segundos'))
```

```r
#Peso vehiculos 
mtcars <- mtcars %>%
  mutate(Peso_kilos=(wt/2)*1000)
```

```r
mtcars <- mtcars %>%
  mutate(Peso = ifelse(Peso_kilos <= 1500,
                       'Livianos',
                       'Pesados'))
```

```r
View(mtcars)
```

### Viendo más información con facet wrap - Parte 1


```r
mtcars %>%
  arrange(desc(Peso))
```

```r
Mas_pesados <- mtcars %>%
  filter(model %in% c('Lincon Continental', 'Chrysler Imperial', 'Cadillac Fleetwood', 'Merc 450SE'))
Mas_pesados
```

```r
ggplot(Mas_pesados, aes(x=hp, y=mpg))+
  geom_point()+
  facet_wrap(~ model)
```

```r
ggplot(mtcars, aes(x=cyl, y=mpg, size=Peso))+
  geom_point()+
  facet_wrap(~am)
```

```r
ggplot(mtcars, aes(x = cyl, y = mpg, size = Peso_kilos, color = factor(gear))) +
  geom_point(alpha = 0.7) +
  facet_wrap(~am) +
  scale_color_brewer(palette = "Set2") +  # Paleta de colores mejorada
  labs(
    x = "Número de cilindros",
    y = "Millas por galón (mpg)",
    size = "Peso_kilos)",
    color = "Número de marchas",
    title = "Relación entre cilindros, eficiencia y peso de los vehículos"
  ) +
  theme_minimal() +  # Estilo minimalista
  theme(
    plot.title = element_text(hjust = 0.5, size = 14, face = "bold"),
    legend.position = "top",
    legend.title = element_text(face = "bold")
  )
```

### Viendo más información con facet wrap - Parte 2 Orangeec

```r
TopNaranjas <- orangeec %>%
  filter(Country %in% c('Mexico', 'Panama', 'Argentina', 'Colombia', 'Brazil', 'Paraguay'))
TopNaranjas

names(orangeec)
```

```r
ggplot(TopNaranjas, aes(x=`Education invest % GDP`,
                        y=`Services % GDP`, size=`GDP PC`))+
  geom_point()+
  facet_wrap(~`Country`)
```

```r
ggplot(TopNaranjas, aes(x=`Internet penetration % population`,
                        y=`Creat Ind % GDP`, size=`Unemployment`))+
  geom_point()+
  facet_wrap(~`Country`)
library(RColorBrewer)
myColors <- brewer.pal(9,'Reds')
```

```r
ggplot(TopNaranjas, aes(x= `Internet penetration % population`,
                        y=`GDP PC`, fill = `Creat Ind % GDP`))+
  geom_tile()+
  facet_wrap(~`Country`)+
  scale_fill_gradientn(colors=myColors)
```

```r
# Suponiendo que myColors es un vector de colores predefinidos
myColors <- c("blue", "green", "yellow", "red")

ggplot(TopNaranjas, aes(x = `Internet penetration % population`,
                        y = `GDP PC`, fill = `Creat Ind % GDP`)) +
  geom_point(shape = 21, color = "black") +  # El color negro para el borde de los puntos
  facet_wrap(~ `Country`) +
  scale_fill_gradientn(colors = myColors) +
  labs(
    x = "Penetración de Internet (% población)",
    y = "PIB per cápita",
    fill = "% Industria Creativa del PIB",
    title = "Relación entre penetración de Internet y PIB per cápita en países de LATAM"
  ) +
  theme_minimal() +
  theme(
    plot.title = element_text(hjust = 0.5, size = 14, face = "bold"),
    legend.position = "right"
  )
```
